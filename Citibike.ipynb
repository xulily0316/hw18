```python
import pandas as pd
```

```python
data18 = ['JC-201801-citibike-tripdata.csv', 'JC-201802-citibike-tripdata.csv', 'JC-201803-citibike-tripdata.csv', 'JC-201804-citibike-tripdata.csv', 'JC-201805-citibike-tripdata.csv', 'JC-201806-citibike-tripdata.csv', 'JC-201807-citibike-tripdata.csv', 'JC-201808-citibike-tripdata.csv', 'JC-201809-citibike-tripdata.csv', 'JC-201810-citibike-tripdata.csv', 'JC-201811-citibike-tripdata.csv', 'JC-201812-citibike-tripdata.csv']

# Create an empty list to store each dataframe
df18_list = []

# Loop through each file in the list
for data in data18:
    df18 = pd.read_csv(data)

    # Append the dataframe to the list
    df18_list.append(df18)

# Concatenate all dataframes in the list into one dataframe
df18 = pd.concat(df18_list)

df18
```

```python
data19 = ['JC-201901-citibike-tripdata.csv', 'JC-201902-citibike-tripdata.csv', 'JC-201903-citibike-tripdata.csv', 'JC-201904-citibike-tripdata.csv', 'JC-201905-citibike-tripdata.csv', 'JC-201906-citibike-tripdata.csv', 'JC-201907-citibike-tripdata.csv', 'JC-201908-citibike-tripdata.csv', 'JC-201909-citibike-tripdata.csv', 'JC-201910-citibike-tripdata.csv', 'JC-201911-citibike-tripdata.csv', 'JC-201912-citibike-tripdata.csv']

# Create an empty list to store each dataframe
df19_list = []

# Loop through each file in the list
for data in data19:
    df19 = pd.read_csv(data)

    # Append the dataframe to the list
    df19_list.append(df19)

# Concatenate all dataframes in the list into one DataFrame
df19 = pd.concat(df19_list)

df19
```

```python
data20 = ['JC-202001-citibike-tripdata.csv', 'JC-202002-citibike-tripdata.csv', 'JC-202003-citibike-tripdata.csv', 'JC-202004-citibike-tripdata.csv', 'JC-202005-citibike-tripdata.csv', 'JC-202006-citibike-tripdata.csv', 'JC-202007-citibike-tripdata.csv', 'JC-202008-citibike-tripdata.csv', 'JC-202009-citibike-tripdata.csv', 'JC-202010-citibike-tripdata.csv', 'JC-202011-citibike-tripdata.csv', 'JC-202012-citibike-tripdata.csv']

# Create an empty list to store each dataframe
df20_list = []

# Loop through each file in the list
for data in data20:
    df20 = pd.read_csv(data)

    # Append the dataframe to the list
    df20_list.append(df20)

# Concatenate all dataframes in the list into one DataFrame
df20 = pd.concat(df20_list)

df20
```

```python
combined_df = pd.concat([df18, df19, df20], axis=0)

combined_df
```

```python
mapping = {0: 'Unknown', 1: 'Male', 2: 'Female'}

combined_df['gender'] = combined_df['gender'].map(mapping)

combined_df
```

```python
combined_df.to_csv('citibike.csv', index=False)
```
